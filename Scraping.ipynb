```python
import requests
from bs4 import BeautifulSoup

# Function to fetch links from a given page URL
def fetch_links_from_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        page_content = response.text
        soup = BeautifulSoup(page_content, 'html.parser')
        page_content_div = soup.find('div', class_='page-content')
        if page_content_div:
            container_div = page_content_div.find('div', class_='container')
            if container_div:
                row_div = container_div.find('div', class_='row')
                if row_div:
                    unique_links = set()
                    for div in row_div.find_all('div', recursive=False):
                        a_tag = div.find('a', href=True)
                        if a_tag:
                            unique_links.add(a_tag['href'])
                    return unique_links
    return set()

# Function to extract the movie name and link from each movie page
def fetch_movie_details(movie_url):
    response = requests.get(movie_url)
    if response.status_code == 200:
        page_content = response.text
        soup = BeautifulSoup(page_content, 'html.parser')

        # Extract movie name
        name = None
        head_box_div = soup.find('div', class_='head-box')
        if head_box_div:
            container_div = head_box_div.find('div', class_='container')
            if container_div:
                media_title_div = container_div.find('div', class_='media-title')
                if media_title_div:
                    h1_tag = media_title_div.find('h1')
                    if h1_tag:
                        name = h1_tag.text.strip()

        # Extract the secondary button link
        link = None
        a_tag = soup.find('a', class_='btn labeled secondary', href=True)
        if a_tag:
            link = a_tag['href']

        return name, link
    return None, None

# List to store all movie names and links
all_movies = []

# First page URL
first_page_url = 'https://web.animerco.org/trending/?get=movies'

# Fetch and process the first page
print(f"Fetching first page: {first_page_url}")
links = fetch_links_from_page(first_page_url)
for link in links:
    movie_name, movie_link = fetch_movie_details(link)
    if movie_name and movie_link:
        all_movies.append({
            'name': movie_name,
            'link': movie_link
        })

# Loop through pages 2 to 10 and collect unique links
base_url = 'https://web.animerco.org/trending/page/'

for page_number in range(2, 10):
    url = f"{base_url}{page_number}/?get=movies"
    print(f"Fetching page: {url}")
    links = fetch_links_from_page(url)

    for link in links:
        movie_name, movie_link = fetch_movie_details(link)
        if movie_name and movie_link:
            all_movies.append({
                'name': movie_name,
                'link': movie_link
            })

# Output the collected movie names and links
print(f"Total movies collected: {len(all_movies)}")
for movie in all_movies:
    print(f"Name: {movie['name']}, Link: {movie['link']}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import base64
# Function to extract the secondary large button link from each movie page
def fetch_large_button_link(movie_url):
    response = requests.get(movie_url)
    if response.status_code == 200:
        page_content = response.text

        soup = BeautifulSoup(page_content, 'html.parser')

        # Find the link in the a tag with class "btn secondary fluid large"
        a_tag = soup.find('a', class_='btn secondary fluid large', href=True)
        if a_tag:
            return a_tag['data-url']
    return None



# Create a new list to store the updated movie dictionaries
updated_movies = []

# For each movie in all_movies, fetch the large button link and update the dictionary
for movie in all_movies:
    large_button_link = fetch_large_button_link(movie['link'])
    decoded_url = base64.b64decode(large_button_link).decode('utf-8')
    updated_movies.append({
        'name': movie['name'],
        'link': movie['link'],
        'download': decoded_url
    })

# # Convert the updated_movies list to a DataFrame
all_movies2 = pd.DataFrame(updated_movies)

all_movies2

```
